```python
# EDA
import pandas as pd
import numpy as np
```

```python

#  Load the data
data = pd.read_csv('../data/synthetic_retail.csv')
```

```python

# 2 Basic info
print("=== Data Info ===")
print(data.info())  # Check column types and missing values
print("\n=== Head of Data ===")
print(data.head())

# Check for missing values
print("\n=== Missing Values ===")
print(data.isnull().sum())
```

```python

# Convert Date to datetime if not already
data['Date'] = pd.to_datetime(data['Date'])

#Basic statistics for numeric columns
print("\n=== Statistics ===")
print(data.describe())

#  Check for unrealistic values
# SalesAmount, LaborHours, InventoryCost, FootTraffic
print("\n=== Checking Ranges ===")
for col in ['SalesAmount', 'LaborHours', 'InventoryCost', 'FootTraffic']:
    print(f"{col} - min: {data[col].min()}, max: {data[col].max()}")

# Optional: Remove any negative or zero values if present
for col in ['SalesAmount', 'LaborHours', 'InventoryCost', 'FootTraffic']:
    data = data[data[col] > 0]
```

```python

# Optional: Save cleaned data
data.to_csv('../data/synthetic_retail_cleaned.csv', index=False)
print("\n Data cleaned and saved to ../data/synthetic_retail_cleaned.csv")
print(data.isnull().sum())
data['SalesAmount'] = data['SalesAmount'].astype(float)
data['LaborHours'] = data['LaborHours'].astype(float)
data['InventoryCost'] = data['InventoryCost'].astype(float)
data['FootTraffic'] = data['FootTraffic'].astype(float)
# Calculate KPIs
data['LaborProductivity'] = data['SalesAmount'] / data['LaborHours']
data['InventoryEfficiency'] = data['SalesAmount'] / data['InventoryCost']
data['FootTrafficEfficiency'] = data['SalesAmount'] / data['FootTraffic']
```

```python

# Quick check
print(data[['LaborProductivity','InventoryEfficiency','FootTrafficEfficiency']].head())
data.head()
```
